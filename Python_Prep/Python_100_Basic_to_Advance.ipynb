Problem 4:

Find the factorial of a number using a loop

Input = 5

5! = 5 x 4 x 3 x 2 x 1 = 120

Output = 120

```python
a = int(input("Enter a Number:"))

if a < 0:
  print("Number is negative")
else:
  fact = 1
  for i in range(2, a+1):
    fact *= i
  print(f"Factorial of {a} is {fact}")
```

Problem 7:

Check if a string is a palindrome

Palindrome = reads same forward and backward

Input: "madam"

Output: Palindrome

```python
string = "madam"
rev_string = ""

for ch in string:
  rev_string = ch + rev_string

print("Palindrome" if string == rev_string else "Not Palindrome")
```

```python
string = "madam"
is_palindrome = True

for i in range(len(string)//2):
  if string[i] != string[-i-1]:
    is_palindrome = False
    break

print("Palindrome" if is_palindrome else "Not Palindrome")
```

```python
string = "madam"
print("Palindrome" if string == string[::-1] else "Not Palindrome")
```

Problem 10: Check if a Number is Prime

Write a program that checks whether a given number is prime or not.

```python
n = int(input("Enter a Number: "))

if n <= 1:
  print("Number is not prime")
else:
  for i in range(2,n):
    if n % i == 0:
      print("Number is not prime")
      break
  else:
      print("Number is prime")
```

Problem 19:

Create a simple calculator that performs:

Addition

Subtraction

Multiplication

Division

```python
num1 = float(input("Enter a number : "))
op = input("Enter a operator(+,-,*,/) : ")
num2 = float(input("Enter a number : "))

if op == "+":
  result = num1 + num2
elif op == "-":
  result = num1 - num2
elif op == "*":
  result = num1 * num2
elif op == "/":
  result = num1 / num2 if num2 != 0 else "Cannot divide by zero"
else:
  result = "Invalid Operator"

print(result)
```

```python
num1 = float(input("Enter a number : "))
op = input("Enter a operator(+,-,*,/) : ")
num2 = float(input("Enter a number : "))

operations = {
    "+" : num1 + num2,
    "-" : num1 - num2,
    "*" : num1 * num2,
    "/" : num1 / num2 if num2 != 0 else "Cannot divide by zero"
}

print(operations.get(op, "Invalid Operator"))
```

Problem 22:

Check Armstrong Number (3-digit).

A number is Armstrong if sum of cubes of its digits = the number itself

```python
number = input("Enter a number : ")
sum = 0

for digit in number:
  sum += int(digit)**3

if int(number) == sum:
  print("It's a Armstrong Number")
else:
  print("It's not a Armstrong Number")
```

```python
number = int(input("Enter a number : "))
original = number
cube_total = 0

while number>0:
  digit = number % 10
  cube_total += digit**3
  number = number // 10

if original == cube_total:
  print("It's a Armstrong Number")
else:
  print("It's not a Armstrong Number")
```

Problem 28:

Convert a decimal (base 10) number to binary (base 2).

```python
n = int(input("Enter a decimal number: "))
binary = ""

if  n == 0:
  binary = "0"
else:
  while n > 0:
    remainder = n % 2
    binary = str(remainder) + binary
    n = n//2

print("binary :", binary)
```

Problem 29:

Convert a binary number (base 2) into decimal (base 10).

```python
binary = input("Enter a binary number: ")
decimal = 0
power = 0

for digit in reversed(binary):
    decimal += int(digit) * (2 ** power)
    power += 1

print("Decimal:", decimal)
```
